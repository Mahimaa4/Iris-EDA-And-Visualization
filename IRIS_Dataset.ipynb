```python
#HISTOGRAM,PDF,CDF
# 1-D SCATTER PLOT OF PETAL LENGTH
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset("iris")   # <--- This is important!

# Separate species
iris_setosa = iris.loc[iris["species"] == "setosa"]
iris_virginica = iris.loc[iris["species"] == "virginica"]
iris_versicolor = iris.loc[iris["species"] == "versicolor"]

# -------- 1-D SCATTER PLOT --------
plt.figure(figsize=(8, 4))
plt.scatter(iris_setosa["petal_length"], [0]*len(iris_setosa), color="blue", label="Setosa")
plt.scatter(iris_versicolor["petal_length"], [1]*len(iris_versicolor), color="green", label="Versicolor")
plt.scatter(iris_virginica["petal_length"], [2]*len(iris_virginica), color="red", label="Virginica")

plt.title("1-D Scatter Plot of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.yticks([0, 1, 2], ["Setosa", "Versicolor", "Virginica"])
plt.legend()
plt.show()
```

```python
#CUMULATIVE DISTRIBUTION FUNCTION

#plot pdf and cdf
import numpy as np
counts,bin_edges=np.histogram(iris_setosa['petal_length'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)

#compute CDF
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example: Compare petal_length for all 3 species
species = {
    "Setosa": iris_setosa["petal_length"],
    "Versicolor": iris_versicolor["petal_length"],
    "Virginica": iris_virginica["petal_length"]
}

plt.figure(figsize=(8,6))

for name, data in species.items():
    # Compute histogram (PDF)
    counts, bin_edges = np.histogram(data, bins=10, density=True)
    pdf = counts / sum(counts)
    
    # Compute CDF
    cdf = np.cumsum(pdf)
    
    # Plot PDF and CDF (different styles)
    plt.plot(bin_edges[1:], pdf, marker='o', linestyle='-', label=f"{name} PDF")
    plt.plot(bin_edges[1:], cdf, marker='x', linestyle='--', label=f"{name} CDF")

plt.xlabel("Petal length")
plt.ylabel("Probability / Cumulative Probability")
plt.title("PDF & CDF of Petal Length for Iris Species")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# MEAN, VARIENCE AND STANDARD VARIATION

print("Mean:")#mean=.average(central tendency)
print(np.mean(iris_setosa["petal_length"]))
print(np.mean(np.append(iris_setosa["petal_length"],50))) #outlier value (by mistake)
print(np.mean(iris_virginica["petal_length"]))
print(np.mean(iris_versicolor["petal_length"]))
```

```python
#Standard deviation  (spread --observe the graph (width of curve))

print("Std-dev")
print(np.std(iris_setosa["petal_length"]))
print(np.std(iris_virginica["petal_length"]))
print(np.std(iris_versicolor["petal_length"]))
```

```python
#MEDIAN (TO HANDLE OUTLIER VALUES WHICH WERE CREATING ERROR IN MEAN)

print("Median")
print(np.median(iris_setosa["petal_length"]))
print(np.median(np.append(iris_setosa["petal_length"],50))) #outlier value (by mistake)
print(np.median(iris_virginica["petal_length"]))
print(np.median(iris_versicolor["petal_length"]))
```

```python
#QUANTILES

print("Quantiles")
print(np.percentile(iris_setosa["petal_length"],np.arange(0,100,25)))
print(np.percentile(iris_virginica["petal_length"],np.arange(0,100,25)))
print(np.percentile(iris_versicolor["petal_length"],np.arange(0,100,25)))

# 50th percentile is median 
# 25th - min value
# 100th - max value
```

```python
print("90th percentile")
print(np.percentile(iris_setosa["petal_length"],90))
print(np.percentile(iris_virginica["petal_length"],90))
print(np.percentile(iris_versicolor["petal_length"],90))
```

```python
# MEDIAN ABSOLUTE DEVIATION                         #Median-->Mean
# IQR-Inter Quartile Range                          #MAD--> Std-Dev
from statsmodels import robust
print("Median absolute deviation")
print(robust.mad(iris_setosa["petal_length"]))
print(robust.mad(iris_virginica["petal_length"]))
print(robust.mad(iris_versicolor["petal_length"]))
```
